## Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#to resolve unicode decode error
file = '../input/moviescsv/movies.csv'
import chardet
with open(file, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
result
df = pd.read_csv(file, encoding='ISO-8859-1')
```

## Let's look at the data


```python
df.head()
```

### See if there's any missing data

```python
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(percent_missing*100)))
```

### Identify data types

```python
df.dtypes
```

### Change data type of columns

```python
df['budget']=df['budget'].astype('int64')
df['gross'] =df['gross'].astype('int64')
```

### Create correct Year column

```python
df['yearcorrect']=df['released'].astype(str).str[0:4]
```

```python
 pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

### Drop any duplicates

```python
df.drop_duplicates()
df.head() # to avoid overloading 
```

## Scatter plot with budget vs Gross

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budge vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

## Let's start looking at correlation

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized


df.head()
```


### Assigns a random numeric value for each unique categorical value

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
